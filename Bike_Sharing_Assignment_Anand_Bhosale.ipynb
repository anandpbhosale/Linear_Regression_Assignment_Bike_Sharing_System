```python
numeric_cols = ["temp", "atemp", "hum", "windspeed", "cnt"]

# Calculating IQR for each column
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)

IQR = Q3 - Q1

# Threshold for Outlier
threshold = 1.5

# Identifying Outlier for each field
outliers = ((df[numeric_cols] < (Q1 - threshold * IQR)) | (df[numeric_cols] > (Q3 + threshold * IQR))).any()
outlier_cols = outliers[outliers].index.tolist()
print("Columns having Outliers are :", outlier_cols)
```

```python
# Selecting the Categorical Fields for Analysis
col1 = "season"
col2 = "weathersit"

''' Creating cross-tabulation [Cross tabulation is typically used 
when you have information that can be divided into mutually exclusive groups, also known as categorical variables.
It allows you to examine relationships within the data that may not be readily apparent.]'''
cross_tab = pd.crosstab(df[col1], df[col2])

# Calculating % (column-wise)
totals = cross_tab.sum(axis=0)
cross_tab_percent = cross_tab.div(totals, axis=1) * 100

# Plotting the stacked bar plot including %
ax = cross_tab_percent.plot(kind="bar", stacked=True)

for i in ax.patches:
    width, height = i.get_width(), i.get_height()
    x, y = i.get_xy()
    ax.annotate(f'{height:.1f}%', (x + width/2, y + height/2), ha = "center", va = "center")
    
plt.xlabel(col1)
plt.ylabel("Percentage")
plt.title("Bivariate Analysis :" + col1 + " vs " + col2)
plt.legend(title = col2)
plt.show()
```

        5.3 Creating Dummy Variables
        </span>
    </font>
<?h3>

- By creating dummy variables, we convert categorical variables into a binary numerical representation. Each category of the variable is transformed into a separate binary variable (0 or 1), making it easier for algorithms to process and interpret the data.

```python
df.columns
```

```python
# Converting the variables to object type to facilitate the creation of Dummy Variables
df["season"] = df["season"].astype(object)
df["mnth"] = df["mnth"].astype(object)
df["weekday"] = df["weekday"].astype(object)
df["weathersit"] = df["weathersit"].astype(object)
```

```python
df.info()
```

```python
# Converting Categorical Variables to Dummy Variables
season_dummy = pd.get_dummies(df["season"], prefix="season", drop_first=True)
mnth_dummy = pd.get_dummies(df["mnth"], drop_first=True)
weekday_dummy = pd.get_dummies(df["weekday"], drop_first=True)
weathersit_dummy = pd.get_dummies(df["weathersit"], prefix="weathersit", drop_first=True)
```

```python
# Adding/Concatenating Dummy Variables with Original Dataset
df = pd.concat([df, season_dummy, weathersit_dummy,weekday_dummy,mnth_dummy], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# Dropping Original Categorical Variables

df.drop(["season", "weathersit", "weekday", "mnth"], axis=1, inplace=True)
```
